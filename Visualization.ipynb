```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Line Graph

```python
np.random.seed(0)
x=list(range(1,13))
y=np.random.randint(500,5000,12)
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,color='red',linestyle='-.',marker='*',markersize=10,linewidth=4,label='Line1')
plt.xlabel('Months') #for x-axis label
plt.ylabel('Count') #for y axis label
plt.title('Month wise Covid Count')
plt.grid() #for background boxes
plt.legend() #to display the label
plt.show()
```

```python
x=np.arange(0,10,0.1)
y=np.sin(x)
plt.plot(x,y,marker='o',color='green',label='sine')
plt.legend()
plt.show()
```

# Bar Graph

```python
x=['Male','Female']
y=[5,10]
plt.bar(x,y,color=['green','red'])
plt.show()
```

```python
df1=pd.DataFrame({'City':list('MDMDCCMMDD'),'Gender':list('MMMMMFFFMF'),
                'Marks 1':np.random.randint(1,101,10)})
df1['City'].replace({'M':'Mumbai','C':'Chennai','D':'Delhi'},inplace=True) 
df1['Gender'].replace({'M':'Male','F':'Female'},inplace=True)
df1.head()#will give top five values
```

```python
df1['Gender'].value_counts()
```

```python
x=df1['Gender'].value_counts().index
y=df1['Gender'].value_counts().values
print(x)
print(y)
plt.bar(x,y,color='green')
plt.show()
```

```python
#using seaborn
sns.countplot(data=df1,x='Gender')
plt.show()
```

```python
#bar using pandas
df1['Gender'].value_counts().plot(kind='bar')
plt.show()
```

```python
#bar using pandas kind change
df1['Gender'].value_counts().plot(kind='pie')
plt.show()
```

# Pie Chart

```python
x=['Rent','Fuel','Food','Miss']
value=[12000,4000,5000,1000]
```

```python
plt.pie(value,labels=x,autopct='%i%%')
plt.show()
```

```python
plt.pie(value,labels=x,autopct='%0.2f%%',explode=[0,0,0.1,0]) #explode is for bring specific value out
plt.show()
```

```python
df1['Gender'].value_counts().plot(kind='pie',autopct='%i%%')
plt.show()
#autopct for percentage 
```

# Histogram

```python
d1=pd.DataFrame({'col1':np.random.randn(1000)})
d1.head()
```

```python
plt.hist(d1['col1'],color='yellow',edgecolor='red')
plt.show()
```

```python
plt.hist(d1['col1'],color='yellow',edgecolor='red',bins=5) #bins is for number of intervals
plt.show()
```

```python
plt.hist(d1['col1'],color='yellow',edgecolor='red',bins=[-4,0,2,5])
plt.show()
```

```python
sns.kdeplot(d1['col1'])
plt.show()
```

```python
sns.kdeplot(d1['col1'],cumulative=True) #cumulative will add values with previous
plt.show()
```

```python
sns.histplot(data=d1,x='col1',kde=True)
plt.show()
```

# Scatterplot

```python
x=[1,2,3,4,5]
y=[5,10,40,5,10]
```

```python
plt.scatter(x,y)
plt.show()
```

```python
df1=pd.DataFrame({'City':list('MDMDCCMMDD'),'Gender':list('MMMMMFFFMF'),
                'Marks 1':np.random.randint(1,101,10)})
df1['City'].replace({'M':'Mumbai','C':'Chennai','D':'Delhi'},inplace=True) 
df1['Gender'].replace({'M':'Male','F':'Female'},inplace=True)
df1['Marks 2']=np.random.randint(1,90,10)
df1.head()
```

```python
plt.scatter(df1['Marks 1'],df1['Marks 2'])
plt.show()
```

```python
sns.scatterplot(data=df1,x='Marks 1',y='Marks 2')
plt.show()
```

```python
sns.scatterplot(data=df1,x='Marks 1',y='Marks 2',hue='Gender')
plt.show()
```

# Jointplot

```python
sns.jointplot(data=df1,x='Marks 1',y='Marks 2')
plt.show()
```

# Lmplot

```python
sns.lmplot(data=df1,x='Marks 1',y='Marks 2')
plt.show()
```

# Stripplot

```python
sns.stripplot(data=df1,x='Gender',y='Marks 1') #allows overlapping of data
plt.plot()
```

# Swarmplot

```python
sns.swarmplot(data=df1,x='Gender',y='Marks 1') #does not allow overlapping of data 
plt.show()
```

# Pairplot

```python
sns.pairplot(df1)
plt.show()
```
