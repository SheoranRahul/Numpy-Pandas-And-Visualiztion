# Duplicate

```python
df.duplicated(keep='first') #top to bottom
```

```python
df.duplicated(keep='last') #bottom to top
```

```python
df.duplicated(['Col1'],keep=False)
```

```python
df.drop_duplicates(keep=False)
```

```python
df.drop_duplicates(['Col1'],keep=False)
```

# Pivot and CrossTab

```python
df1=DataFrame({'City':list('MDMDCCMMDD'),'Gender':list('MMMMMFFFMF'),
                'Marks 1':np.random.randint(1,101,10)})
df1['City'].replace({'M':'Mumbai','C':'Chennai','D':'Delhi'},inplace=True) 
df1['Gender'].replace({'M':'Male','F':'Female'},inplace=True)
df1
```

```python
pd.crosstab(df1['City'],df1['Gender'])
```

```python
pd.crosstab(df1['City'],df1['Gender'],margins=True)
```

```python
df1.pivot_table(values='Marks 1',index='City',columns='Gender')
```

```python
df1.pivot_table(values='Marks 1',index='City',columns='Gender',aggfunc=['min','max'])
```

# Concatenation

```python
df1=DataFrame(np.arange(1,10).reshape(3,3),columns=list('ABC'),index=[1,2,3])
df2=DataFrame(np.arange(1,13).reshape(4,3),columns=list('ABC'),index=[4,5,6,7])
print(df1)
df2
```

```python
pd.concat([df1,df2],keys=['A','B'])
```

```python
df1=DataFrame(np.arange(1,10).reshape(3,3),columns=list('ABC'),index=[1,2,3])
df2=DataFrame(np.arange(1,13).reshape(4,3),columns=list('DEF'),index=[1,2,3,4])
print(df1)
df2
```

```python
pd.concat([df1,df2],axis=1)
```

```python
pd.concat([df1[['A','B']],df2[['D','E']]],axis=1)
```

# Merge

```python
df1=DataFrame({'ID':[101,102,103,101],'M1':[10,20,30,40]})
df2=DataFrame({'ID':[104,102,101],'M2':[90,10,50]})
df1
```

```python
pd.merge(df1,df2,how='outer',on='ID')
```

```python
pd.merge(df1,df2,how='left',on='ID')
```
